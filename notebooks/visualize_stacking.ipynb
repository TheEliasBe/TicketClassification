```python
import pandas as pd
import matplotlib.pyplot as plt
import tikzplotlib

stacking_df = pd.read_csv("stacked_2.csv")
stacking_df = stacking_df[["before", "after"]]
stacking_df = stacking_df[stacking_df["before"] > 0]
stacking_df.describe()
```

```python
# perform a kolmogorov-smirnov test to check if the distributions are the same
from scipy.stats import ks_2samp
ks_2samp(stacking_df["before"], stacking_df["after"], alternative="greater")
```

```python

fig = stacking_df.boxplot().get_figure()

plt.xlabel("Stacking")
plt.ylabel("Accuracy")
# plt.title("Accuracy Improvement after applying Stacking")

# save as svg
fig.savefig("plots/stacking_improvement_2.svg")
tikzplotlib.save("plots/stacking_improvement_2.tex")
fig
```
